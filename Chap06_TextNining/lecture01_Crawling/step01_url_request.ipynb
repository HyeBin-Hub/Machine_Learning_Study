```python
"""
news 제공 사이트 
url : http://media.daum.net
> pip install beautifulsoup4
"""
import requests # url 요청 
from bs4 import BeautifulSoup # source -> html 파싱 

# 1. url 요청 
url = "http://media.daum.net"
req = requests.get(url)
print(req) # <Response [200]>
```

```python
src = req.text # source
print(src)
```

```python
# 2. html 파싱 : html 문서 생성 
html = BeautifulSoup(src, "html.parser")
print(html) # <tag>
```

```python
# 3. tag[속성='값'] 추출 
a_tag = html.select("a[class='link_txt']") # tag 수집 
print(len(a_tag)) # 75
print(type(a_tag)) # <class 'list'>
# [원소1, 원소2, ....]
a_tag
```

```python
# 4. tag 내용 추출 -> 문서 수집 
crawling_data = [] # 수집 문서 저장 

cnt = 0
for a in a_tag : 
    #print(a.string)
    cont = str(a.string) # tag 내용 
    # 문서 끝 불용어(줄바꿈,공백,특수문자) 제거 -> list 저장 
    crawling_data.append(cont.strip()) 
    if '코스피' in crawling_data :
        print(cnt) # 59
    cnt += 1 
```
