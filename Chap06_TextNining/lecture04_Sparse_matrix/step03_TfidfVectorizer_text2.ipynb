```python
"""
sparse matrix 작업 절차 
 1. csv file load : temp_spam_data2.csv
 2. texts(x), target(y) 전처리 
 3. max features 
 4. sparse matrix
"""

import string # text 전처리 
# sparse matrix
from sklearn.feature_extraction.text import TfidfVectorizer 
import pandas as pd # csv file 


# 1. csv file load
spam_data = pd.read_csv("C:/Users/hyebin/Desktop/study/python_ML/data/textMining_data/temp_spam_data2.csv",
            header = None, encoding='utf-8')
spam_data.info()
```

```python
# 2. texts(x), target(y) 전처리 
target = spam_data[0] # 1번칼럼 
texts = spam_data[1] # 2번칼럼 

print(target) # ham(0) vs spam(1)
print(texts)
```

```python
# 1) target -> dummy 생성 
target = [1 if t == 'spam' else 0 for t in target]
target [:50]# [0, 1, 0, 1, 0]
```

```python
# 2) texts 전처리 : 숫자, 특수문자, 공백 제거 
def text_prepro(texts):
    # Lower case
    texts = [x.lower() for x in texts]
    # Remove punctuation
    texts = [''.join(c for c in x if c not in string.punctuation) for x in texts]
    # Remove numbers
    texts = [''.join(c for c in x if c not in string.digits) for x in texts]
    # Trim extra whitespace
    texts = [' '.join(x.split()) for x in texts]
    return texts

# 함수 호출 : texts 전처리 
texts = text_prepro(texts)   
print(texts[:5])
```

```python
# 3. max features 
'''
tfidf = TfidfVectorizer()
tfidf_fit = tfidf.fit(texts) # texts -> words 
'''
tfidf_fit = TfidfVectorizer().fit(texts)
voc = tfidf_fit.vocabulary_ # 단어 반환 
print(voc)
len(voc) # 8603
```

```python
#max_features = len(voc) # 8603
max_features = 4000  
'''
max_features = 4000 : 절반 단어를 이용해서 분석  
'''
```

```python
# 4. sparse matrix
tfidf = TfidfVectorizer(max_features=max_features, stop_words='english')
sparse_mat = tfidf.fit_transform(texts) # stances -> sparse matrix
print(sparse_mat.shape) # (5574, 8336) -> (5574, 4000)
print(sparse_mat)
```

```python
arr_sparse_mat = sparse_mat.toarray()
print(arr_sparse_mat,"\n")

print(arr_sparse_mat[:20, 100:150])
arr_sparse_mat.shape # (5574, 4000)
```

```python
# 5. train/test split 
from sklearn.model_selection import train_test_split
train_texts, test_texts, train_target, test_target = train_test_split(arr_sparse_mat, target, test_size = 0.3)

print(train_texts.shape) # (3901, 4000)
print(test_texts.shape) # (1673, 4000)
```

```python
# 6. pickle file save 
import numpy as np 

spam_train_test = (train_texts,test_texts,train_target ,test_target )  # 튜플로 묵어줌
np.save("C:/Users/hyebin/Desktop/lee/Machine Learning_Study/data/spam_train_test.npy",spam_train_test )
```
