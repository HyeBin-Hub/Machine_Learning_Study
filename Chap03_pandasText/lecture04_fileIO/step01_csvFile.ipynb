```python
# 특수문자 -> NaN 처리(NA) 
conv = {'키' : '$', '몸무게' : '-'}
student_txt3 = pd.read_table('C:/Users/hyebin/Desktop/Python_ML/data/student2.txt', sep='\s+',
                             na_values = conv)
student_txt3
```

```python
h = student_txt3['키']
w = student_txt3['몸무게']

print('키의 평균 =', h.mean())
print('가장 큰 키 =', h.max())
print('몸의 평균 =', w.mean())
```

```python
height = h.mean()
weight = round(w.mean(), 2)
print(weight)
```

```python
# 칼럼 추가 
student_txt3['hmean'] = height
student_txt3['wmean'] = weight
```

```python
# 3. 파일 저장 

# 파일 저장 
student_txt3.to_csv('C:/Users/hyebin/Desktop/Python_ML/data/student_output.csv', 
                    index=None, na_rep='NaN', encoding='utf-8')

student_df = pd.read_csv('C:/Users/hyebin/Desktop/Python_ML/data/student_output.csv', encoding='utf-8')
student_df
```
