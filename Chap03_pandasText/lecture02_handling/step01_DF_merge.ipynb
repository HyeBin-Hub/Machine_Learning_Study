```python
'''
data frame 병합(합치기) 
'''

from pandas import Series
import pandas as pd
```

```python
# 1. Series Merge
s1 = Series([1,3], index=['a','b'])
s1
```

```python
s2 = Series([5,6,7], index=['a','b','c'])
s2
```

```python
s3 = Series([11,13], index=['a','b']);s3
```

```python
# 행 단위 결합 
result = pd.concat([s1, s2, s3], axis = 0) # 행 단위 결합 
print(result)
print(type(result)) # Series
```

```python
# 열 단위 결합 : DF 생성 
result2 = pd.concat([s1, s2, s3], axis = 1) # 열 단위(index) 결합 
print(result2)
print(type(result2)) # DataFrame
```

```python
# 2. DataFrame Merge
wdbc = pd.read_csv('./data/wdbc_data.csv')
print(wdbc.info())
```

```python
cols = wdbc.columns 
cols
```

```python
df1 = wdbc[cols[0:16]] # 1~16
df1
```

```python
sid = wdbc.id # Series
sid
```

```python
df2 = wdbc[cols[16:]] # 17~32
df2['id'] = sid # id 칼럼 추가 
print(df2.columns)
```

```python
# id 칼럼 공통 포함
print(df1)
```

```python
print(df2)
```

```python
# 1. id 칼럼으로 DF 병합

# merge
# 외부조인 : how='outer' -> 왼쪽 df 기준 join
df_merge = pd.merge(df1, df2) # how='inner'
print(df_merge.info()) # id 칼럼 기준 
```

```python
print(df_merge) # id 칼럼 기준
```

```python
# 2. 칼럼단위 df 붙이기 : 32칼럼 합치기 
df1 = wdbc[cols[:16]] # 1~16
df1
```

```python
df2 = wdbc[cols[16:]] # 17~32
df2
```

```python
# concat
df_con = pd.concat([df1, df2], axis = 1) # 열 단위(index) 결합 
print(df_con)
```
