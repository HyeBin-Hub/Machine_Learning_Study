```python
'''
iris.csv
 - 산점도 matrix 시각화 
'''

from pandas import plotting

# iris 산점도 matrix
iris = pd.read_csv('C:/Users/hyebin/Desktop/study/python_ML/data/iris.csv')
cols = list(iris.columns) # 칼럼 추출 
```

```python
col4 = iris[cols[:4]]
#pd.scatter_matrix(col4)
plotting.scatter_matrix(col4)
plt.show()
```

```python
# iris 3d 산점도
from mpl_toolkits.mplot3d import Axes3D

col_x = iris[cols[0]]
col_y = iris[cols[1]]
col_z = iris[cols[2]]

species = []
for s in iris['Species'] :
    if s == 'setosa' :
        species.append(1)
    elif s == 'versicolor' :
        species.append(2) 
    else :
        species.append(3)
        
print(species) 
```

```python
fig = plt.figure()
chart = fig.add_subplot(1,1,1, projection='3d')

chart.scatter(col_x, col_y, col_z, c=species)
chart.set_xlabel('Sepal.Length')
chart.set_ylabel('Sepal.Width')
chart.set_zlabel('Petal.Length')
#plt.legend(loc='best') # 1234
plt.show()
```
