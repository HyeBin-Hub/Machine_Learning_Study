```python
'''
marker, color, line style, label 이용 
'''

import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot') # 차트 격차 제공 
```

```python
# 난수 누적합
plot_data1 = np.random.randn(50).cumsum()
plot_data2 = np.random.randn(50).cumsum()
plot_data3 = np.random.randn(50).cumsum()
plot_data4 = np.random.randn(50).cumsum()

fig = plt.figure()
chart = fig.add_subplot(1,1,1)
chart.plot(plot_data1, marker=r'o', color=u'blue', linestyle='-', label='Blue Solid')
chart.plot(plot_data2, marker=r'+', color=u'red', linestyle='--', label='Red Dashed')
chart.plot(plot_data3, marker=r'*', color=u'green', linestyle='-.', label='Green Dash Dot')
chart.plot(plot_data4, marker=r's', color=u'orange', linestyle=':', label='Orange Dotted')
chart.xaxis.set_ticks_position('bottom')
chart.yaxis.set_ticks_position('left')

chart.set_title('Line Plots: Markers, Colors, and Line styles')
plt.xlabel('index')
plt.ylabel('Random Number')
plt.legend(loc='best')
plt.show()
```
