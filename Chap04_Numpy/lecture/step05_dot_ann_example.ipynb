```python
"""
신경망에서 행렬곱 예
 - 은닉층(h) = [입력(x) * 가중치(w)] + 편향(b)
"""
import numpy as np

# ANN model 예

'''
input : image(28,28) * weight(28*32) = hidden node(28*32)
'''

# x data 생성 : image(28,28)
x_img = np.random.randint(0, 2, 784) # 1차원(0~1 난수 정수) 
x_img2d = x_img.reshape(28, 28) # 2d
print(x_img2d.shape) # (28, 28)
x_img2d # 0 or 1 난수 정수 
```

```python
# weight data 생성 
weight = np.random.randn(28, 32) # 표준정규분포(평균=0, 편차=1)
print(weight[:5])
weight.shape # (28, 32)
```

```python
# hidden node 생성 
hidden = np.dot(x_img2d, weight)
print(hidden[:5])
hidden.shape #(28, 32) x_img2d (행 :28,  weight열 :32)
```
