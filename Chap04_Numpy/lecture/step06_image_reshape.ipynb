```python
"""
image 규격화(reshape)
"""

from glob import glob # 파일 검색 패턴 사용(*.jpg)
from PIL import Image # image file open
import numpy as np 
import matplotlib.pyplot as plt # image 출력 

path = "C:/Users/hyebin/Desktop/study/python_ML/images"
file = path + '/test2.jpg'
```

```python
'''
PIL image vs numpy image
PIL.resize() : image size 가능 
numpy.resize() 없음
'''

# PIL 객체 이미지 로드 
img = Image.open(file)
print(type(img)) # <class 'PIL.JpegImagePlugin.JpegImageFile'>
print(np.shape(img)) # (360, 540, 3)
plt.imshow(img) # 이미지 출력 
img
```

```python
# PIL object -> numpy object형 변환
arrImg = np.asarray(img)
print(arrImg.shape) # (360, 540, 3), (332, 250, 3)
print(arrImg.max()) # 255 색상중 가장 밝은색 
print(arrImg.min()) # 0
arrImg# 이미지의 픽셀에 대한 내용을 볼수 있다
```

```python
# 모든 image 규격화 함수 정의 
def imageReshape() :
    # image reshape size
    img_h = 150
    img_w = 150
    
    image_reshape = [] # image 저장 
    
    for file in glob(path + '/*.jpg') : # jpg 파일 탐색 
        # image file 읽기
        img = Image.open(file)  
        
        # image resize
        img_re = img.resize( (img_w, img_h)) # 크기 변경 
        # PIL -> numpy
        img_data = np.asarray(img_re)
        
        # image 저장 
        image_reshape.append(img_data)
        
        print(file, ":", img_data.shape)
        '''
        D:/Python_ML/chap04_Numpy/images\test1.jpg : (150, 150, 3)
        D:/Python_ML/chap04_Numpy/images\test2.jpg : (150, 150, 3)
        '''
    return np.array(image_reshape) # list(1) -> numpy(n array)
```

```python
# 함수 호출 
image_reshape = imageReshape()  
print(image_reshape.shape)
```
